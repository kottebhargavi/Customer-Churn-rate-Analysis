```python
df.fillna(0, inplace=True)      # Fill with 0
df.dropna(inplace=True)         # Remove null rows
```

```python
df['Churn']
```

```python
df['Churn'] = df['Churn'].map({'Yes':1, 'No':0})
```

```python
df
```

```python
df['Churn']
```

```python
df.groupby('Contract')['Churn'].mean()
```

```python
df['ChargeGroup'] = pd.cut(df['MonthlyCharges'],
                          bins=[0,30,60,90,120],
                          labels=['Low','Medium','High','Very High'])
```

```python
df.groupby('ChargeGroup')['Churn'].mean()
```

```python
df['TenureGroup'] = pd.cut(df['tenure'],
                          bins=[0,12,24,48,72],
                          labels=['0-1yr','1-2yr','2-4yr','4-6yr'])
```

```python
df.groupby('TenureGroup')['Churn'].mean()
```
